```python
import os
from zipfile import ZipFile

# Define o nome da pasta a ser criada e o nome do arquivo .zip
nome_pasta = 'IRECE'
nome_arquivo_desejado_no_zip = 'nome_original.txt'

for i in range(2003,2023):
  nome_zip = f'{i}.zip'
  nome_arquivo_desejado = f'INMET_NE_BA_A424_IRECE_01-01-{i}_A_31-12-{i}.CSV'
  novo_nome_arquivo = f'INMET_IRECE_{i}.txt'
  # Cria a pasta
  caminho_pasta = f'/content/drive/MyDrive/TCC_WIND/{nome_pasta}'
  os.makedirs(caminho_pasta, exist_ok=True)

  # Define o caminho completo do arquivo .zip
  caminho_zip = f'/content/drive/MyDrive/TCC_WIND/{nome_zip}'

  try:
    # Extrai apenas o arquivo desejado do arquivo .zip
    with ZipFile(caminho_zip, 'r') as zip_ref:
        zip_ref.extract(nome_arquivo_desejado, caminho_pasta)
        caminho_original = os.path.join(caminho_pasta, nome_arquivo_desejado)
        caminho_novo = os.path.join(caminho_pasta, novo_nome_arquivo)
        os.rename(caminho_original, caminho_novo)
    # Exibe a lista de arquivos na pasta após a extração
    print(f'Arquivo {nome_arquivo_desejado} extraído em {caminho_pasta}.')
  except:
    print(f'Arquivo {nome_arquivo_desejado} não existe em {nome_zip}.')
```
